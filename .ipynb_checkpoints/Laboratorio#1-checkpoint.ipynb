# Laboratorio #1

## Que es Git?

Git es proyecto de codigo abierto y con mantenimiento activo. Es un sistema para el manejo de versiones, el mas utilizado en el mundo.

Git permite editar documentos, comunmente codigo y mantener una version de estos y registro de los cambios especificos realizados en cada version.

## Que es GitHub?

GitHub es un sitio web para alojar codigo de cualquier desarrollador utilizando la tecnologia Git.

GitHub permite que los desarrolladores subas su codigo mientras estos sean de codigo abierto, si se desea que el proyecto no sea de codigo abierto, hay una opcion de paga para esto.
