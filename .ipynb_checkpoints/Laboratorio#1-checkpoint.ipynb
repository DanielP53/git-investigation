## Que es Git?

Git es proyecto de codigo abierto y con mantenimiento activo. Es un sistema para el manejo de versiones m√°s utilizado en el mundo. Permite editar documentos, comunmente codigo y mantener una version de estos y registro de los cambios especificos realizados en cada version.

Utiliza una arquitectura distribuida donde el historial de versiones no se mantiene en un solo lugar, las copias de trabajo de cada desarrollador tambien es un repositorio.



### Funcionalidad

Git funciona principalmente con tres estados:
- Working Directory:mar
    Son los archivos locales, donde comunmente se realizan modificaciones a los archivos.
        
- Staging Area:
    Son loas archivos/cambios que alcanzan un milestone o un punto donde queremos generar una version. No por generar una version significa que se puedan revertir,modificar o sobreescribir estos cambios.
    
- Repository:
    Comunmente es almacenado en la nube y funciona comunmente como fuente de verdad para los repositorios locales.

El repositorio es la unica etapa de Git que no sucede localmente, debido a que multiples desarrolladores pueden necesitar acceso desde distintas maquinas, localidades y horas.

### Manipulacion

Git tiene herramientas visuales que pueden "facilitar" el manejo de versiones, pero tambien es muy facil y hay extensa documentacion de como utilizarlo por medio de comandos en la terminal.

Los comandos mas basicos son:

- git clone __url__, este se utiliza para clonar de forma local, url es la direccion de donde se extrae el repositorio.

- git add __filepath__, con este comando se agrega el archivo al staging area.

- git commit -m "__message__", sirve para crear un "paquete" con todos los cambios que se encuentran en el staging area con un mensaje descriptivo de los cambios revisados, para despues subirlos al repositorio

- git push -u __origin__, este comnado subira el paquete creado con el commit al repositorio

## Que es GitHub?

GitHub es un sitio web para alojar codigo de cualquier desarrollador utilizando la tecnologia Git.

GitHub permite que los desarrolladores subas su codigo mientras estos sean de codigo abierto, si se desea que el proyecto no sea de codigo abierto, hay una opcion de paga para esto.



Al ser los proyectos de codigo abierto permite a otros usuarios contribuir, mejorar y comentar los proyectos. Tambien permite que los usuarios esten seguros de cual es la funcionalidad de la aplicacion y que no tenga codigo malicioso.

GitHub tiene tantos proyectos de codigo que ha desarrollado distintas tecnologias para apoyar a sus usuarios. El mas conocido el Copilot, es una inteligencia artificial que da sugerencias a los desarrolladores, les dice posibles errores antes de que ejecuten el codigo y mas.
